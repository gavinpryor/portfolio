```python
from textblob import TextBlob # nip tasks
from newspaper import Article # scrape + extract info
```

```python
url = 'https://medium.com/@davidsweenor/what-is-analytics-maturity-and-why-does-it-matter-e3c18d430ccb'

article = Article(url)

article.download()
article.parse()
article.nlp()

text = article.summary

blob = TextBlob(text)

sentiment = blob.sentiment.polarity # result between 1 to -1

print(sentiment)
```
